```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
spotify_songs = pd.read_csv('Popular_Spotify_Songs.csv', encoding='ISO-8859-1')
```

```python
spotify_songs.value_counts()
```

```python
spotify_songs.value_counts(['in_spotify_playlists'])
```

```python
spotify_songs.value_counts(['track_name'])
```

```python
# first few rows of the dataframe
print("first  rows of the dataframe:")
print(spotify_songs.head())

# statistics of the numerical columns
print("\n statistics of the numerical columns:")
print(spotify_songs.describe())

# Information about the dataframe
print("\ninformation about the dataframe:")
print(spotify_songs.info())

# check for missing values
print("\nmissing values in the dataframe:")
print(spotify_songs.isnull().sum())

#  the distribution of songs by released year
print("\ndistribution of songs by released year:")
print(spotify_songs['released_year'].value_counts())

#  the correlation between numerical features
print("\ncorrelation matrix of numerical features:")
print(spotify_songs.corr())

#  Top 10 most streamed songs
top_10_songs = spotify_songs[['track_name', 'artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(10)
print("\ntop 10 most streamed songs:")
print(top_10_songs)

#  Average BPMc of songs by year
average_bpm_byyear = spotify_songs.groupby('released_year')['bpm'].mean()
print("\naverage BPM of songs by year:")
print(average_bpm_byyear)
"""
# Save  data to a new CSV file
output_file = 'analyzed_spotify_data.csv'
spotify_songs.to_csv(output_file, index=False)
print(f"\ data saved to {output_file}")"""
```

```python
year = spotify_songs['released_year'].value_counts()
# Créer un graphique
plt.figure(figsize=(10,6))
year.plot(kind='bar') #le type de representation
plt.title('Distribution des annes de realise')
plt.xlabel('annes')
plt.ylabel('nombre de chanson')
plt.xticks(rotation=90, ha='right')
plt.grid(True)
plt.tight_layout() # la mise en page du graphique pour le assure que le graphe est correctement agencé
plt.show() #affiche le graphe
```

```python
top_10_songs = spotify_songs[['track_name', 'artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(10)
print("\ntop 10 most streamed songs:")
print(top_10_songs)
```

```python
average_bpm_byyear = spotify_songs.groupby('released_year')['bpm'].mean()
print("\naverage BPM of songs by year:")
print(average_bpm_byyear)
```

```python
# Distribution of songs by released year
data = spotify_songs.copy()

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='released_year', palette='viridis')
plt.title('Distribution of Songs by Released Year')
plt.xlabel('Released Year')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()

# Visualization: Correlation matrix heatmap
plt.figure(figsize=(12, 8))
datan = data.select_dtypes(include=['float64', 'int64'])
corr_m = datan.corr()
sns.heatmap(corr_m, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Visualization: Top 10 most streamed songs
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_songs, x='streams', y='track_name', palette='inferno')
plt.title('Top 10 Most Streamed Songs')
plt.xlabel('Streams')
plt.ylabel('Track Name')
plt.show()

# Visualization: Average BPM of songs by year
plt.figure(figsize=(10, 6))
average_bpm_byyear.plot(kind='bar', color='dodgerblue')
plt.title('Average BPM of Songs by Year')
plt.xlabel('Released Year')
plt.ylabel('Average BPM')
plt.xticks(rotation=45)
plt.show()

# Additional visualizations
# Example: Distribution of Danceability
plt.figure(figsize=(10, 6))
sns.histplot(data['danceability_%'], kde=True, color='skyblue')
plt.title('Distribution of Danceability')
plt.xlabel('Danceability (%)')
plt.ylabel('Frequency')
plt.show()

# Example: Relationship between Energy and Valence
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='energy_%', y='valence_%', hue='released_year', palette='coolwarm')
plt.title('Relationship between Energy and Valence')
plt.xlabel('Energy (%)')
plt.ylabel('Valence (%)')
plt.legend(title='Released Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
