```python
import pandas as pd
import matplotlib.pyplot as plt

spotify_songs = pd.read_csv('Popular_Spotify_Songs.csv', encoding='ISO-8859-1')


# Afficher les premières lignes des DataFrames pour vérifier le chargement
print("spotify song title Dataset:")
print(spotify_songs.head())
```

```python
spotify_songs.value_counts()
```

```python
spotify_songs.value_counts(['in_spotify_playlists'])
```

```python
spotify_songs.value_counts(['track_name'])
```

```python
year = spotify_songs['released_year'].value_counts()
# Créer un graphique
plt.figure(figsize=(10,6))
year.plot(kind='bar') #le type de representation
plt.title('Distribution des annes de realise')
plt.xlabel('annes')
plt.ylabel('nombre de chanson')
plt.xticks(rotation=50, ha='right')
plt.grid(True)
plt.tight_layout() # la mise en page du graphique pour le assure que le graphe est correctement agencé
plt.show() #affiche le graphe
```
